# Formating Data

```python
import numpy
data['pixels'] = data['pixels'].map(lambda x: np.array(x.split(), dtype = 'float32').reshape(48,48))
data.head()
```

```python
images = np.array(data['pixels'].to_list())
```

```python
labels = data['gender'].values
```

```python
images_reshape = images.reshape(images.shape[0], 48, 48, 1)
```

```python
images
```

```python
if np.max(images_reshape) > 1:
  images_reshape = images_reshape/255.0
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(images_reshape, labels, test_size = 0.2, random_state = 42)
```

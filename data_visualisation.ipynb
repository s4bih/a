```python
# Sample data
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
sales = [15000, 18000, 20500, 22000, 19500, 25000, 28000, 24500, 23000, 26000, 30000, 32500]
```

```python
plt.plot(months,sales,marker='o',color='r')
plt.xticks(rotation=90)

plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales')

plt.show()
```

```python
plt.bar(months,sales)
plt.xticks(rotation=45)
plt.tight_layout()
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales')

plt.show()
```

```python
categories = ['Housing', 'Utilities', 'Transportation', 'Groceries', 'Entertainment', 'Healthcare', 'Education', 'Miscellaneous']

expenses = [1200, 300, 400, 500, 200, 150, 250, 100]
```

```python
plt.bar(categories,expenses)
plt.xticks(rotation=45)
plt.tight_layout()
plt.xlabel('Categories')
plt.ylabel('Expenses')
plt.title('Expense Analysis')

plt.show()
```

```python
plt.plot(categories,expenses,marker='o')
plt.xticks(rotation=45)
plt.tight_layout()
plt.xlabel('Categories')
plt.ylabel('Expenses')
plt.title('Expense Analysis')

plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(categories,expenses)
plt.xticks(rotation=45)
plt.tight_layout()
plt.xlabel('Categories')
plt.ylabel('Expenses')
plt.title('Expense Analysis')

plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.stem(categories,expenses)
plt.xticks(rotation=45)
plt.tight_layout()
plt.xlabel('Categories')
plt.ylabel('Expenses')
plt.title('Expense Analysis')

plt.show()
```
